```python
#add rank to countries for each year
df2.sort_values("hdi_value", ascending=False, inplace=True)

def ranker(df):
    df['hdi_rank'] = np.arange(len(df)) + 1
    return df

df2 = df2.groupby(['year']).apply(ranker)
df_hdi_with_rank = df2.groupby("year").apply(pd.DataFrame.sort_values, "countries")

for index, country in df_fgi.iterrows():
    df_hdi_with_rank.loc[(df_hdi_with_rank.year == country.year) & (df_hdi_with_rank.countries == country.countries), "fgi_rank"] = country.fgi_rank
    df_hdi_with_rank.loc[(df_hdi_with_rank.year == country.year) & (df_hdi_with_rank.countries == country.countries), "fgi_value"] = country.fgi_value

for index, country in df_hfi.iterrows():
    df_hdi_with_rank.loc[(df_hdi_with_rank.iso == country.iso) & (df_hdi_with_rank.year == str(country.year)), "hfi_rank"] = country.hfi_rank
    df_hdi_with_rank.loc[(df_hdi_with_rank.iso == country.iso) & (df_hdi_with_rank.year == str(country.year)), "hfi_value"] = country.hfi_value
     
df_hdi_with_rank.head()
```
